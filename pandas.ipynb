# Subsetting and sorting a dataframe

```python
#to sort the individuals column,
homelessness.sort_values("individuals") #it automatically sorts in ascending order
```

```python
#to sort in descending order,
homelessness.sort_values("individuals", ascending=False)
```

```python
#subsetting based on  text data
#Let's say we are filtering only south Atlanta on the region column
homelessness[homelessness["region"]=="South Atlantic"]
```

```python
#we can subset based on multiple conditions
homelessness[homelessness["region"]=="South Atlantic"]
homelessness[homelessness["state"]=="Virginia"]
#homelessness[(homelessness["region"]=="South Atlantic") & (homelessness["state"]=="Virginia")]
 
```
